```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn import preprocessing
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv ('data.csv',encoding='utf-8-sig')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('CA', axis=1, inplace=True)
df.drop('year', axis=1, inplace=True)
df.drop('month', axis=1, inplace=True)
df.drop('median_dom', axis=1, inplace=True)
```

```python
df.head()
```

```python
y = df['quality']
X = df.drop('quality',axis=1)
#, x_test, y_train, y_test = train_test_split(X, y, test_size=0.70, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.10, random_state=1)
```

```python
svmLinearKernelClassifier = svm.LinearSVC();
svmLinearKernelClassifier.fit(X_train, y_train)
```

```python
# Test Set
print("-----------------------Test Accuracy - Poly Kernel-----------------------")
y_pred = svmLinearKernelClassifier.predict(X_test)
print(classification_report(y_test, y_pred))


# Validation Set
print("-----------------------Validation Accuracy - Poly Kernel-----------------------")
y_val_pred = svmLinearKernelClassifier.predict(X_val)
print(classification_report(y_val, y_val_pred))

#Check Train Accuracy
print("-----------------------Train Accuracy - Poly Kernel-----------------------")
y_train_pred = svmLinearKernelClassifier.predict(X_train)
print(classification_report(y_train, y_train_pred))
```
